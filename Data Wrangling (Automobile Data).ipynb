```python
%matplotlib inline
plt.hist(df["horsepower"])
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Without Bins")
```

```python
bins = np.linspace(min(df["horsepower"]),max(df["horsepower"]),4)
bins
```

```python
group_name = ["Low","Medium","High"]
```

```python
df["horsepower-binned"] = pd.cut(df["horsepower"], bins, labels=group_name, include_lowest=True)
df[["horsepower","horsepower-binned"]].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
plt.bar(group_name, df["horsepower-binned"].value_counts())
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower Bins")
```

```python
plt.hist(df["horsepower"], bins = 3)
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("Horsepower Bins using 'bins' atrribute")
```
